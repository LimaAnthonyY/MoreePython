## O que é o Matplotlib?


**Matplotlib** é um das principais bibliotecas Python para visualização de dados para a criação de gráficos 2D.

> **matplotlib.pyplot** é um módulo do Matplotlib para a plotagem de gráficos de forma semelhante ao MATLAB.

![picture](https://matplotlib.org/stable/_images/sphx_glr_anatomy_001.png)

```python
# Plotagem simples via numpy
x = np.arange(0, 2 * math.pi, 0.05)
y = np.sin(x)
print (x)
plt.plot(x,y)
```

```python
# Plotagem simples via numpy com elementos no gráfico
plt.xlabel("ângulo")        # rótulo do eixo X
plt.ylabel("seno")          # rótulo do eixo Y
plt.title('Onda senoidal')  # título do gráfico
plt.plot(x,y)
plt.show()
```

## Figuras

O objeto `Figure` é um contêiner de alto nível utilizado para abranger todos os elementos de plotagem.

```python
# Dados para plotagem
x1 = np.linspace(1, 100, 10)
x2 = np.linspace(1, 200, 10)
y = np.linspace(100, 5000, 10)
```

```python
# Uso básico
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])            # [left, bottom, width, height]
linha1 = ax.plot(x1, y)                 # (x1, y)
linha2 = ax.plot(x2, y)                 # (x2, y)
plt.show()
```

```python
# Parâmetros do Figure
fig = plt.figure(figsize=(10, 3),     # (width,height) em polegadas
                 dpi=72,              # Pontos por polegada (define qualidade; para plotar na web, 72dpi é bom)
                 facecolor='green',   # Cor de fundo
                 edgecolor='red',     # Cor da borda
                 linewidth=20)        # Espessura da borda
ax = fig.add_axes([0,0,1,1])
linha1 = ax.plot(x1, y)
plt.show()
```

```python
# Plotagem com elementos
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
linha1 = ax.plot(x1, y)
linha2 = ax.plot(x2, y)
ax.legend(labels = ('Smartphones', 'Televisores'), loc = 'lower right')
ax.set_title("Efeito da propaganda nas vendas")
ax.set_xlabel('medium')
ax.set_ylabel('vendas')
plt.show()
```

* Hist;
* hist2d (histograma 2D);
* Pie (gráfico de pizza);

## Multiplots e subplots

```python
x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.tanh(x)
```

***Multiple plots***

```python
plt.plot(x, y1, color='g', label='seno')
plt.plot(x, y2, color='b', label='cosseno')
plt.plot(x, y3, color='r', label='tangente')
plt.plot(x, y4, color='y', label='tangente hiperbólica')
plt.xlabel("Ângulo")
plt.ylabel("Magnitude")
plt.legend()
plt.show()
```

***Subplots***

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(5, 5))
ax[0, 0].plot(x, y1)
ax[0, 0].set_title("Seno", fontsize=10)

ax[0, 1].plot(x, y2)
ax[0, 1].set_title("Cosseno", fontsize=10)

ax[1, 0].plot(x, y3)
ax[1, 0].set_title("Tangente", fontsize=10)

ax[1, 1].plot(x, y4)
ax[1, 1].set_title("Tangente hiperbólica", fontsize=10)

fig.tight_layout()  # ajuste automático dos gráficos à área selecionada

plt.savefig("grafico.pdf")
plt.show()
```
