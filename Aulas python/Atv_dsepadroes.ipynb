```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from apyori import apriori
```

```python
df = pd.read_csv('retail_dataset.csv', sep=',', header=0)
```

```python
df.info()
```

```python
# Converte a base de dados em listas de listas
registros = []

for i in range(0, df.shape[0]):
    registros.append([str(df.values[i, j]) for j in range(0, df.shape[1])])

registros[0:1]
```

```python
regras = apriori(registros, min_support = 0.003, min_confidence = 0.3, min_lift = 4, min_length = 2)

lista_regras = list(regras)

print(f"# Regras: {len(lista_regras)}")
```

```python
print(lista_regras[1])
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0]) for result in results]
    rhs         = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
```

```python
df_resultados = pd.DataFrame(inspect(lista_regras),
                             columns = ['Antecedente', 'Consequente', 'Suporte', 'Confiança', 'Lift'])

df_resultados.nlargest(n = 10, columns = 'Lift')
```

```python
df_resultados.nsmallest(n = 10, columns = 'Lift')
```
